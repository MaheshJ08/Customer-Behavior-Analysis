```python
inputs="""SELECT * 
FROM custpro.geography AS geo
RIGHT JOIN (
    SELECT 
        a.CustomerID,
        a.Gender,
        a.Age,
        a.GeographyID,
        b.ProductID,
        b.Rating,
        b.ReviewText
    FROM custpro.customer_info AS a
    LEFT JOIN custpro.cu_reveiw AS b
        ON a.CustomerID = b.CustomerID
) AS cu
ON geo.GeographyID = cu.GeographyID;"""

cur.execute(inputs)
var=cur.fetchall()
ouputs=pd.DataFrame(var,columns=cur.column_names)
```
